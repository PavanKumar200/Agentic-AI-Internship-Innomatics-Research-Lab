```python
#User Login Check
username = 'admin' #initialize 
password = '1234'

uname = input("Enter Username :") #user inputs
upass = input("Enter Password: ")

if(uname == username and upass == password):  #condition check
    print(f"Hello {uname}! Login Successful!")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
PassStudent = 0 #initialize
FailStudent = 0
for mark in marks: #condition check
    if mark > 50:
        PassStudent += 1
    else:
        FailStudent += 1
        
print(f"Total No of Pass students = {PassStudent}") #print output
print(f"Total No of Fail students = {FailStudent}")
      
```

```python
#Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
res= [] #initialize
for name in names:
    res.append(name.strip().lower())
print(res)
```

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

for m in messages:
    length = len(m)
    print(f"Length of {m} :{length} ") #length of string 
    
    if length > 10:
        print(f"Flagged this '{m}': Message is too long!") #flag the message
    
```

```python
# Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0 #initialize
for i in logs:
    if i.lower() == 'error':
        error_count +=1 #count errors
print(f"Total number of errors in logs are : {error_count}")
```
