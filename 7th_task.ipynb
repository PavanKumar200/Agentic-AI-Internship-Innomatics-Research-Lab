```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(f"{self.name} is {self.status}")



light_name = "Bedroom Light"
action = "ON"

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
else:
    light.turn_off()


light.show_status()
```

```python
#Problem 2: Employee ID Card System
class employee_ID:
    def __init__ (self,name,emp_id,dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def show_details(self):
        print("Employee ID Card")
        print(f"Name : {self.name}")
        print(f"ID : {self.emp_id}")
        print(f"Departmant : {self.dept}")

emp1 = employee_ID("Arun","E8929","Sales")

emp1.show_details()

        
```

```python
#Problem 3: Mobile Contact Record
class Contact_Number_Save:
    def __init__(self,name,number):
        self.name = name
        self.number = number

    def Saved(self):
        print("Contact Saved")
        print(f"Name : {self.name} \nContact : {self.number}")

contact1 = Contact_Number_Save("Kumar",8794568795)

contact1.Saved()
```

```python
#Problem 4: Product Price Tag Generator
class price_tag_generator:
    def __init__(self,prdt_name,price):
        self.prdt_name = prdt_name
        self.price = price
    def show_price_tag(self):
        print(f"Product : {self.prdt_name}\nPrice: ₹{self.price}")
prdt1 = price_tag_generator("Shampoo",6799)
prdt1.show_price_tag()
```

```python
#Problem 5: Movie Rating Display System
class movie_rating:
    def __init__(self,movie_name,rating):
        self.movie_name = movie_name
        self.rating = rating
    def show_ratings(self):
        print(f"Movie : {self.movie_name}\nRating: {self.rating}/5.0")

movie1 = movie_rating("RRR",4.4)
movie1.show_ratings()
        
```

```python
#Delivery Address Manager
class DeliveryAddress:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


customer_name = "Suman"
address = "Hyderabad"


delivery = DeliveryAddress(customer_name, address)


delivery.show_details()
```
