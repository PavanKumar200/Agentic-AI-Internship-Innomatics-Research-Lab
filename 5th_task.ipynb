```python
# Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    # Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Output
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


# Example usage
likes = [150, 200, 300, 400]  # Total = 1050
analyze_post_engagement(likes)
```

```python
#Agriculture – Rainfall Adequacy Checker
def check_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = 0

    # Loop to calculate total rainfall
    for rainfall in rainfall_data:
        total_rainfall += rainfall

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)

    # Determine rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Output
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")


# Example usage
seasonal_rainfall = [70, 75, 68, 75]  # Average = 72
required_rainfall = 70
check_rainfall_adequacy(seasonal_rainfall, required_rainfall)
```

```python
#Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):
    # Convert list to set to remove duplicates
    unique_usernames = set(usernames)

    # Check for duplicates
    if len(unique_usernames) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    # Output
    print(f"Duplicate Accounts Found: {result}")


# Example usage
usernames_list = ["alex", "john", "emma", "alex"]
detect_duplicate_accounts(usernames_list)
```

```python
#Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices, premium_threshold=2000):
    premium_crops = []

    # Loop to filter premium crop prices
    for price in prices:
        if price > premium_threshold:
            premium_crops.append(price)

    # Output
    print(f"Premium Crops: {premium_crops}")


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
#System Monitoring – Application Health Checker
def check_application_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")


# Example usage
errors = 7
check_application_health(errors)
```

```python
#Banking – Daily Transaction Limit Checker
def check_transaction_limit(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")


# Example usage
transaction_amount = 60000
check_transaction_limit(transaction_amount)
```

```python
# E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_records):
    total_classes = len(attendance_records)
    attended_classes = 0

    # Loop to count attended classes
    for status in attendance_records:
        if status == 1:   # 1 represents Present
            attended_classes += 1

    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    # Output
    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {eligibility}")


# Example usage
# 1 = Present, 0 = Absent
attendance = [1, 1, 0, 1, 1]
check_exam_eligibility(attendance)
```

```python
# Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
    bill = 0

    # Slab-wise billing calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example usage
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
```
