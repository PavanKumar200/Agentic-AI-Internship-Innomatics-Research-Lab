```python
#Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]
min_num = numbers[0]
max_num = numbers[0]

for i in numbers:
    #check min number
    if min_num > i:
        min_num = i

    #check max number
    if max_num < i:
        max_num = i
print("Min is :",min_num)
print("Max is :",max_num)
```

```python
#Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count+=1
print("No. products price above 1000 are :",count)
```

```python
#Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]
presents = []
P_count=0
No_of_days = len(attendance)
for present in attendance:
    if present == "P":
        presents.append(present)
        P_count+=1
percentage = P_count/No_of_days*100
print(f"No of days Present in a week: {P_count} days")
print(f"Percentage: {percentage:.2f}")
```
