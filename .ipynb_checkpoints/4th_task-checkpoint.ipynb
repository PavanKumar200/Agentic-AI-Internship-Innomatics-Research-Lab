```python
# Mobile Recharge Validation System
def recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount should be at least 50! TRY AGAIN")
        return False

    if amount in valid_plans:
        print("Recharge Success!")
        return True
    else:
        print("Enter a valid plan")
        return False


while True:
    print("Choose your plan [199, 299, 399, 599]")
    amount = int(input("Enter recharge amount: "))
    if recharge(amount):
        break
```

```python
# Inventory Reorder Alert System
def reorder_alert(stocks):
    for prdt, qty in stocks.items():
        if qty < 15:
            print(f"{prdt} :{qty} Reorder!")
        else:
            print(f"{prdt} : {qty} Stock ok")
    print("Stock check completed")


stocks = {
    "Rice": 2,
    "Wheat": 95,
    "Sugar": 60,
    "Salt": 150,
    "Oil": 4,
    "Milk": 40,
    "Tea": 14,
    "Coffee": 25,
    "Soap": 200,
    "Detergent": 110
}

reorder_alert(stocks)
```

```python
#Student Result Processing System
def result_calculate():
    marks = []
    subjects = int(input("Enter number of students"))
    for i in range(subjects):
        mark = int(input(f"Enter marks of subject {i+1}:"))
        marks.append(mark)
    print(marks)
    total_marks = 0
    for i in marks:
        total_marks+= i
    avg_marks = total_marks/subjects
    print(f"Total marks : {total_marks}")
    print(f"Average marks : {avg_marks}")

    if avg_marks >= 50:
        return "PASS! CONGRATS"
    else:
        return "FAIL! WORK HARD!"

res = result_calculate()
print(res)
```

```python
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = int(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    if distance > 0 and (peak == "yes" or peak == "no"):
        total_fare = calculate_fare(distance, peak)
        print("Total Cab Fare = $", total_fare)
        break
    else:
        print("Invalid input! Try again.\n")
```

```python
#Employee Attendance Eligibility Checker
def attendence_checker():
    days = 7
    attendence = [] #initialize
    #add p or a to list
    for i in range(days):
        attended = input("Enter the attendence of the week of 7 days 'P/p'-present or 'A/a'-absent:").lower()
        attendence.append(attended)
    print(attendence)
    #calculate percentage
    prsnt_days = 0
    for i in attendence:
        if i == 'p':
            prsnt_days +=1
    percentage = prsnt_days / days *100 
    print(f"Percentage: {percentage:.2f}%")
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
            
    

res = attendence_checker()
print(res)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # final validation
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password "


pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
