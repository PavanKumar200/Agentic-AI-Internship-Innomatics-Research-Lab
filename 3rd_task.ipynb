```python
# Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
scores = []
emp_names = []

#append values to list
for key,value in employees.items():
    scores.append(value)

#check scores
for key,value in employees.items():
    max_score = max(scores)
    #check the same values
    if max_score == value:
        emp_names.append(key)

print(f"Top Performers Eligible for Bonus: {emp_names} (Score : {max_score})")
    
```

```python
#Search Query Keyword Analysis
search_query = "Buy mobile phone buy, phone online".lower()

punct = ".,!?;:'\""
clean_text = ""

for ch in search_query:
    if ch not in punct:
        clean_text += ch

words = clean_text.split()


freq ={}
res = {}
for word in words:
    if word in freq:
        freq[word] +=1
    else:
        freq[word] = 1

for word in freq:
    if freq[word] > 1:
        res[word] = freq[word]

print(res)
```

```python
#Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

domain_cnt = {}

for mail in emails:
    domain = mail.split('@')[1]

    if domain in domain_cnt:
        domain_cnt[domain]+=1
    else:
        domain_cnt[domain] = 1
total = len(emails)

for domain in domain_cnt:
    percentage = (domain_cnt[domain] *100 / total)
    print(domain,":",int(percentage),"%")
```
