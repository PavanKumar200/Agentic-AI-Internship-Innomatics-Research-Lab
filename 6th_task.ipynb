```python
def check_parking(capacity, logs):
    vehicles_in = 0

    for log in logs:
        if log == 'IN':
            if vehicles_in < capacity:
                vehicles_in += 1
            else:
                print("Parking is full. Vehicle cannot enter.")
        elif log == 'OUT':
            if vehicles_in > 0:
                vehicles_in -= 1
            else:
                print("Parking is empty. No vehicle to exit.")

    available_parking = capacity - vehicles_in

    print("\n--- Parking Status ---")
    print("Currently Parked vehicles:", vehicles_in)
    print("Available parking slots:", available_parking)

    if vehicles_in == capacity:
        print("Parking status: FULL")
    else:
        print("Parking Status: AVAILABLE")


parking_capacity = int(input("Enter the parking capacity: "))
vehicle_logs = []

print("Enter parking logs ('IN' / 'OUT'). Type 'exit' to stop.")

while True:
    entry = input().strip().upper()
    if entry == 'EXIT':
        break
    elif entry in ['IN', 'OUT']:
        vehicle_logs.append(entry)
    else:
        print("Invalid input. Enter 'IN', 'OUT', or 'exit'.")

check_parking(parking_capacity, vehicle_logs)
```

```python
def calculate_delivery_time(distance_kms, traffic_level, weather):
    per_kms = 5
    base_time = distance_kms * per_kms

    # traffic conditions
    if traffic_level == 'high':
        base_time = base_time + 10
    elif traffic_level == 'medium':
        base_time = base_time + 5
    elif traffic_level == 'low':
        base_time = base_time + 0

    # weather conditions
    if weather == 'stormy':
        base_time = base_time + 10
    elif weather == 'rainy':
        base_time = base_time + 5
    elif weather == 'cloudy':
        base_time = base_time + 3
    elif weather == 'clear':
        base_time = base_time + 0

    return base_time


Distance_km = int(input("Enter the distance: "))
Traffic_Level = input("Enter the traffic level HIGH/MEDIUM/LOW: ").lower()
Weather = input("Enter the weather status (rainy/cloudy/clear/stormy): ").lower()

result = calculate_delivery_time(Distance_km, Traffic_Level, Weather)

print("Estimated Delivery Time:", result, "mins")
```

```python
import random

def seat_occupancy(booked_seats_list):
    total_seats = len(booked_seats_list)
    booked_seats_count = 0

    for seat in booked_seats_list:
        if seat == 1:
            booked_seats_count += 1

    occupancy = (booked_seats_count / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    if booked_seats_count == total_seats:
        print("Housefull")
    elif occupancy >= 75:
        print("Almost full")
    else:
        print("Not housefull")


seats = int(input("Enter number of seats: "))
booked_seats_list = []

for i in range(seats):
    seat_data = random.randint(0, 1)
    booked_seats_list.append(seat_data)

seat_occupancy(booked_seats_list)
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor
def monitor_resource_usage(resource_usage):
    # Define allowed usage limits (hours)
    usage_limits = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused_resources = []

    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > usage_limits.get(resource, 0):
            overused_resources.append(resource)

    # Display results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resource_usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller
def manage_event_registration(event_capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Process each registration
    for i in range(1, total_registrations + 1):
        if confirmed < event_capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Determine registration status
    if confirmed == event_capacity:
        status = "Closed"
    else:
        status = "Open"

    # Output results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_capacity = int(input("Enter Event Capacity: "))
registrations = int(input("Enter Total Registrations: "))

manage_event_registration(event_capacity, registrations)
```
